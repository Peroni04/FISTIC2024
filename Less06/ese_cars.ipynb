```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv("https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/cars.csv")
df
```

```python
df=df[df["Price"]!="Ask For Price"]
```

```python
df['Price'] = df['Price'].astype(str).str.replace(',', '', regex=False)
```

```python
df['Price'].astype(int)
```

```python
df['fuel_type'] = df.apply(lambda row: row['kms_driven'] if row['kms_driven'] == 'Petrol' else row['fuel_type'], axis=1)
df['kms_driven'] = df['kms_driven'].astype(str).str.replace('Petrol', '', regex=False)
df.dropna(subset=['fuel_type'], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['kms_driven'] = df['kms_driven'].astype(str).str.replace(' kms','', regex=False)
df['kms_driven'] = df['kms_driven'].astype(str).str.replace(',','', regex=False)
```

```python
df['kms_driven'] = df['kms_driven'].replace(['', np.nan], regex=False)
df["kms_driven"].astype(int)
```

```python
X=df.drop(columns="Price")
y=df["Price"]
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3,random_size=42)
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
y_train=ss.fit_transform(y_train)
```
